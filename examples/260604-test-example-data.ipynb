````python
# gemini gave me this helpful codeeee

# 1. Zenodo metadata (where to get it)
RECORD_ID = "6574265"
REMOTE_PATH = "data/UKBB/ssl_data/data.zip"

# 2. Local metadata (where to put it)
# This will save it precisely as 'data.zip' inside a parent directory's 'data/ssl_downstream' folder
LOCAL_DIR = "../data/UKBB/ssl_downstream"
LOCAL_FILENAME = "adl_30hz_clean.zip"

# Clean adjustment to the downloader to accept custom local names
def download_zenodo_to_custom_path(record_id, remote_path, local_dir, local_filename):

    url = f"https://zenodo.org/records/{record_id}/files/{remote_path}?download=1"
    full_local_path = os.path.join(local_dir, local_filename)

    os.makedirs(local_dir, exist_ok=True)

    print(f"Downloading from Zenodo...")
    response = requests.get(url, stream=True)
    response.raise_for_status()

    total_size = int(response.headers.get('content-length', 0))
    with open(full_local_path, 'wb') as f, tqdm(total=total_size, unit='iB', unit_scale=True) as bar:
        for chunk in response.iter_content(chunk_size=8192):
            bar.update(f.write(chunk))

    print(f"Successfully saved to: {full_local_path}")

# Run it
download_zenodo_to_custom_path(RECORD_ID, REMOTE_PATH, LOCAL_DIR, LOCAL_FILENAME)
````
